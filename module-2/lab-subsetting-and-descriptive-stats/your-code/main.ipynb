#### Find the mean and the standard deviation of the temperature of each state.

*Hint: Use functions from Data Manipulation lesson*

```python
# your answer here
NY_mean_t=temp[(temp.State=='New York')].mean()
NY_std_t=temp[(temp.State=='New York')].std()

C_mean_t=temp[(temp.State=='Connecticut')].mean()
C_std_t=temp[(temp.State=='Connecticut')].std()

NJ_mean_t=temp[(temp.State=='New Jersey')].mean()
NJ_std_t=temp[(temp.State=='New Jersey')].std()

print('NY temp:',NY_mean_t,'NY std:',NY_std_t)
print('C temp:',C_mean_t,'C std:',C_std_t)
print('NJ temp:',NJ_mean_t,'NJ std:',NJ_std_t)
```
