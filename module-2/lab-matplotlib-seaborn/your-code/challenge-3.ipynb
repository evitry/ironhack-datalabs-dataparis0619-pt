#### Using Seaborn plot the correlation matrix of various features.

```python
#your code here
'''tit=titanic.select_dtypes(exclude='object')
heat_map = sns.heatmap(tit)
plt.show()'''
```

```python
corr = titanic.corr()
corr
```

```python
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
