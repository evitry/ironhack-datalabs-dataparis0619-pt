````python
table=soup.find_all("div",{'class':'mw-parser-output'})

python_link=[]
for e in table:
    link_tags=e.find_all('ul')
    for link_tag in link_tags:
        link=link_tag.find('li').find('a')
        link2=link.get('href')
        python_link.append(link2)

python_link
````

````python
#Récupération des données
table=soup.find_all("table")[3]
body_tags=table.find("tbody").find_all("tr",limit=20)

date_table=[]
latitude_table=[]
longitude_table=[]
region_table=[]
for row in body_tags:
    dates_tags=row.find('td',{'class':'tabev6'}).find('a')
    for dates_tag in dates_tags:
        date_table.append(re.sub("\xa0\xa0\xa0"," ",dates_tag))
    latitude_tags=row.find_all('td',{'class':'tabev1'})[0]
    latitude_degrees_tags=row.find_all('td',{'class':'tabev2'})[0]
    for latitude_tag in latitude_tags:
        latitude_table.append(latitude_tag)
    longitude_tags=row.find_all('td',{'class':'tabev1'})[1]
    longitude_degrees_tags=row.find_all('td',{'class':'tabev2'})[1]
    for longitude_tag in longitude_tags:
        longitude_table.append(longitude_tag)
    region_tags=row.find('td',{'class':'tb_region'})
    for region_tag in region_tags:
        region_table.append(region_tag)

        
date=pd.Series(date_table)
latitude=pd.Series(latitude_table)
longitude=pd.Series(longitude_table)
region=pd.Series(region_table)
#Création du dataframe
earthquakes=pd.concat([date,latitude,longitude,region],axis=1)
print(earthquakes)
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
#your code
html=requests.get(url).content
soup=BeautifulSoup(html,'lxml')
central_featured_tags=soup.find('div',{'class':'central-featured'}).find_all('a')

wiki_languages_articles=[]
for central_featured_tag in central_featured_tags:
    language=central_featured_tag.find('strong').text
    nb_article=central_featured_tag.find('bdi',{'dir':'ltr'}).text
    wiki_languages_articles.append((language,re.sub('\xa0','',nb_article)))
print(wiki_languages_articles)
#print(central_featured_tags)
````
